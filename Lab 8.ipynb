# Numpy

```python
import numpy as np  #importing library

a=np.array([1,2,3])   #declaring array
b=np.array([4,5,6])
c=np.array([(7,8,9),(10,11,12),(0,1,2)])
  
print(c.shape)  #print the dimension of an array
print(c.size)
print(c)

print(np.add(a,b))  #built-in (add) function for addition
print(np.multiply(a,b))  #built-in (multiply) function for multiply
print(np.divide(a,b))  #built-in (divide) function for divide

print(a.sum())  #array wise sum

print(a.mean())  #average of an array
```
